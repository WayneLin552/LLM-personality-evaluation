```python
import openai
#import newspaper
import json
#from newscatcher import Newscatcher, describe_url,urls
import time
from collections import defaultdict
from tqdm.notebook import tqdm
from concurrent.futures import ThreadPoolExecutor
import concurrent
import pickle
from google.colab import files
import pandas as pd
import numpy as np
#open file manually and change base64.decodestring() into base64.decodebytes(), bugs will be fixed
```

```python
def get_name_url(link):
  name_url = {}
  try:
    paper = Newscatcher(website = link)
    results = paper.get_news()
    articles = results['articles']
  except:
    print(link)
  for article in articles:
    name_url[article['title']]=article['link']
    time.sleep(0.33)
  return name_url
```

```python
#test = get_name_url(urls(language = 'EN')[:2])
name_url = {}
# We can use a with statement to ensure threads are cleaned up promptly
with ThreadPoolExecutor(max_workers=5) as executor:
    # Start the load operations and mark each future with its URL
    future_to_url = {executor.submit(get_name_url, url): url for url in urls(language = 'EN')[-500:]}
    for future in concurrent.futures.as_completed(future_to_url):
        url = future_to_url[future]
        try:
            name_url.update(future.result())
        except Exception as exc:
            print('%r generated an exception: %s' % (url, exc))
```

```python
def get_tot_sum(title,url):
  tot_sum = {}
  try:
    article = newspaper.Article(url=url, language='en')
    article.download()
    article.parse()
    if len(article.text.split())>=50 and len(article.text.split())<=2000:
      tot_sum[title]=article.text.replace('\n\n',' ').replace('\n',' ')
    return tot_sum
  except:
    print(title)
```

```python
#test = get_name_url(urls(language = 'EN')[:2])
tot_sum = {}
# We can use a with statement to ensure threads are cleaned up promptly
with ThreadPoolExecutor(max_workers=5) as executor:
    # Start the load operations and mark each future with its URL
    future_to_url = {executor.submit(get_tot_sum, link,url): url for link, url in name_url.items()}
    for future in concurrent.futures.as_completed(future_to_url):
        url = future_to_url[future]
        try:
            tot_sum.update(future.result())
        except Exception as exc:
            print('%r generated an exception: %s' % (url, exc))
```

```python
with open('tot_sum.pkl','wb') as f:
  pickle.dump(tot_sum, f, protocol=pickle.HIGHEST_PROTOCOL)
files.download('tot_sum.pkl')
#len(tot_sum)
```
