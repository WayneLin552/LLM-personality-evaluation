```python


"""
Put your OpenAI API Key here
"""
openai.api_key = "#########"


def get_response(prompt):
  messages = [
      {"role": "user", "content": prompt}
      ]
  response = openai.ChatCompletion.create(
      model="gpt-3.5-turbo-0613",
      messages=messages,
      temperature=0.8, # set 0.8 to let the response be more random to some extent
      max_tokens=1024
      )
  response_message = response["choices"][0]["message"]
  response_text = response_message['content']
  num_token = response["usage"]
  return response_text, num_token
```

```python
# @title Run OpenAI API and prediction several times for each prompt
from collections import defaultdict
from tqdm.notebook import tqdm


def run_model(prompt_list, iter=3):
  results = defaultdict(list)
  # Predict personality on ChatGPT's response to each prompt N(=iter) times
  for prompt_id, prompt in enumerate(prompt_list):
      for j in tqdm(range(iter)):
          res_text, num_token = get_response(prompt)
          pred = predict(res_text, dataset='kaggle')

          # Store response and # of output tokens just in case
          results['prompt_id'].append(prompt_id)
          results['response'].append(res_text)
          results['num_output_tokens'].append(num_token['completion_tokens'])

          # Concat probabilities into one dict
          for k, v in pred.items():
              results[k].append(v)
  return results

# results = run_model(prompt_list, iter=3)
```

```python
import torch
from transformers import (
    AutoModelForSequenceClassification,
    AutoTokenizer,
    AutoConfig,
)

from peft import PeftModel

# Login to Hugging Face
from huggingface_hub import notebook_login
notebook_login()
```

```python
# Model name/path on Hugging Face Hub
model_name_or_path = 'Yuta555/Llama-2-7b-MBTI-classification'

# Load configration, model, and tokenizer
config = AutoConfig.from_pretrained(model_name_or_path)
model = AutoModelForSequenceClassification.from_pretrained(
    model_name_or_path,
    config=config,
    torch_dtype=torch.bfloat16,
    device_map="auto",
    use_flash_attention_2=True, # speed up inference
)
tokenizer = AutoTokenizer.from_pretrained(model_name_or_path)

# Load the Lora model
model = PeftModel.from_pretrained(model, model_name_or_path)

# Merge the adapter weights with the base model which allows the model's inference to speed up
merged_model = model.merge_and_unload()
```

```python
# Function to prediction using tweets
def predict(tweet: str):
    """
    Input: Tweet texts (str). It should include 50 tweets and be indexed in the format below:
        -----
        1. XXX
        2. XXX
        ...
        -----
    Return: Predicted MBTI label in the form of a four-letter alphabet (str).
    """

    inputs = tokenizer(tweet, return_tensors="pt").to('cuda')

    with torch.no_grad():
        outputs = merged_model(**inputs)
        predictions = outputs.logits.argmax(dim=-1).item()
        predictions = model.config.id2label[predictions]

    return predictions
```

```python
# tweet=list(tweet.items())
len(titles)
without_replacement=[]
```

```python
import random
# without replacement
test = random.sample(titles, 20*50)

for i in range(20):
  temp = test[i*50:i*50+50]
  text = ''
  for j in range(len(temp)):
    if j==0:
      text=text+"1. "+tweet[temp[j]]
    else:
      text=text+"\n"+str(j+1)+". "+tweet[temp[j]]
  without_replacement.append(text)
```

```python
result = []
for dp in without_replacement:
    pred_label = predict(dp)
    result.append(pred_label)
result = pd.DataFrame({'raw':without_replacement,'predict':result})
result
```
