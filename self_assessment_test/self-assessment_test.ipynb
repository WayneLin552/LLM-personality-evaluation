```python
import pandas as pd
import json
from tqdm.notebook import tqdm

from openai import OpenAI
```

```python
client = OpenAI(
    api_key="PUT YOUR OPENAI AP KEY"
)
```

```python
SAVE_PATH = 'llms_mbti_en.json'

mbti_questions = json.load(
    open('mbti_questions_en.json', 'r', encoding='utf8')
)

def get_model_answer(question: str):
    instruction = "Answer the following question according to your most genuine feeling. Just respond with the option number, like: A or B."
    prompt = f"""{instruction}

``
{question}
``"""
    response = client.chat.completions.create(
        model="gpt-3.5-turbo-1106",
        messages=[
            #{"role": "system", "content": "You are a helpful assistant "},
            {"role": "user", "content": prompt}
        ],
        temperature=0,
        max_tokens=5,
    )

    answer = response.choices[0].message.content.strip()

    return answer


# Obtain answers
answers = []
for question in tqdm(mbti_questions.values()):
    res = get_model_answer(question['question'])
    answers.append(res)
```

```python
# Check answers manually so that the first character of them is A or B
answers
```

```python
cur_model_score = {
    'E': 0,
    'I': 0,
    'S': 0,
    'N': 0,
    'T': 0,
    'F': 0,
    'J': 0,
    'P': 0
}

for question, answer in tqdm(zip(mbti_questions.values(), answers)):
    mbti_choice = question[answer[0]]
    cur_model_score[mbti_choice] += 1

e_or_i = 'E' if cur_model_score['E'] > cur_model_score['I'] else 'I'
s_or_n = 'S' if cur_model_score['S'] > cur_model_score['N'] else 'N'
t_or_f = 'T' if cur_model_score['T'] > cur_model_score['F'] else 'F'
j_or_p = 'J' if cur_model_score['J'] > cur_model_score['P'] else 'P'

mbti_res = {
    'details': cur_model_score,
    'res': ''.join([e_or_i, s_or_n, t_or_f, j_or_p])
}

json.dump(mbti_res, open(SAVE_PATH, 'w', encoding='utf8'), indent=4)

print(mbti_res)
print(f'[Done] Result has saved at {SAVE_PATH}.')
```
